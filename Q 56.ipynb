```python
# 56. Generate a generic 2D Gaussian-like array

import numpy as np
import matplotlib.pyplot as plt

# Define the size of the 2D array and standard deviation
size = (100, 100) 
std_dev = 10.0   


x = np.linspace(-size[0] / 2, size[0] / 2, size[0])
y = np.linspace(-size[1] / 2, size[1] / 2, size[1])
x, y = np.meshgrid(x, y)


gaussian_array = np.exp(-(x**2 + y**2) / (2 * std_dev**2))


gaussian_array /= np.max(gaussian_array)

plt.imshow(gaussian_array, cmap='viridis')
plt.colorbar()
plt.title('2D Gaussian-like Array')
plt.show()
```
